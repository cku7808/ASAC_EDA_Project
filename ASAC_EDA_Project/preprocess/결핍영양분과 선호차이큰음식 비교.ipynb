```python
# 선호 차이 큰 음식
prefer_20s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/data_20s_common.csv" 
prefer_30s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/data_30s_common.csv"
```

```python
prefer_20s = pd.read_csv(prefer_20s_path)
prefer_30s = pd.read_csv(prefer_30s_path)
prefer_20s.drop(columns="Unnamed: 0",inplace=True)
prefer_30s.drop(columns="Unnamed: 0",inplace=True)
```

```python
prefer_20s = prefer_20s[prefer_20s['abs(남자-여자 순위 차이)'] > 4]
prefer_20s.sort_values(by="abs(남자-여자 순위 차이)",ascending=False,inplace=True)
save_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/data_20s_common_2.csv"
prefer_20s.to_csv(save_path)
```

```python
prefer_30s = prefer_30s[prefer_30s['abs(남자-여자 순위 차이)'] > 4]
prefer_30s.sort_values(by="abs(남자-여자 순위 차이)",ascending=False,inplace=True)
save_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/data_30s_common_2.csv"
prefer_30s.to_csv(save_path)
```

```python
food_20s = prefer_20s.loc[:,["19-29 음식","남자-여자 순위 차이"]]
food_20s.sort_values(by="남자-여자 순위 차이",inplace=True)
food_20s.set_index("19-29 음식",inplace=True)
```
