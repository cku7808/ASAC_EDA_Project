```python
# 성별 다빈도음식 df 
file_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/data_20s.csv"
data = pd.read_csv(file_path,sep=",")
data.drop(columns="Unnamed: 0",inplace=True)
food_list = data.iloc[:,1].to_list()
len(food_list)
```

```python
path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/rawdata/19-49 (3).xlsx"
food_db = pd.read_excel(path,sheet_name="시트1")
# food_db.drop(index=0, axis=0,inplace=True)
food_db
```

```python
food_db_list = food_db.loc[:,"식품명"]
food_db_list = food_db_list[food_db_list.notnull()]
food_db_list=set(food_db_list)
len(food_db_list)
```

```python
m_20s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/male_data_20s.csv"
f_20s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/female_data_20s.csv"
m_30s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/male_data_30s.csv"
f_30s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/female_data_30s.csv"
```

```python
paths = [m_20s_path,f_20s_path,m_30s_path,f_30s_path]
result = []
for path in paths:
  data = pd.read_csv(path)
  data.drop(columns="Unnamed: 0",inplace=True)
  data.set_index("순위",inplace=True)
  data.replace("도우넛(소없음)","도우넛",inplace=True)
  data.replace("빵(소없음)","빵",inplace=True)
  # print(data)
  for idx,name in enumerate(data.iloc[:,0]):
    if name not in food_db_list:
      data.drop(idx+1,axis=0,inplace=True)
  data.reset_index(drop=True,inplace=True)
  data.insert(0,"순위",range(1,len(data)+1))

  data.to_csv(path)
  result += data.iloc[:,1].to_list()
result = set(result)
```
