```python
import pandas as pd
```

```python
m_20s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/male_data_20s.csv"
f_20s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/female_data_20s.csv"
m_30s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/male_data_30s.csv"
f_30s_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData/female_data_30s.csv"
```

```python
# 데이터 이상치 탐색
paths = [m_20s_path, f_20s_path, m_30s_path, f_30s_path]

for path in paths:
  data = pd.read_csv(path)
  data.drop(columns="Unnamed: 0",inplace=True)
  # print(data)
  Q1 = data.iloc[:,2].quantile(.25)
  Q3 = data.iloc[:,2].quantile(.75)
  IQR = Q3-Q1
  Max = IQR + 1.5*Q3

  tmp = data.iloc[:,2].apply(lambda x: x>Max)
  data["이상값"] = tmp
  data.to_csv(path)
# # 저장 
#   save_path = "/content/drive/MyDrive/Tacademy ASAC/EDA Project/finishedData"
# data_list2 = ["male_data_20s", "female_data_20s", "male_data_30s", "female_data_30s"]

# for i in range(len(data_list)):
#   data_list[i].to_excel(save_path+"/"+data_list2[i]+".xlsx")
```

```python
male_data_20s = pd.read_csv(m_20s_path)
male_data_20s.drop(columns="Unnamed: 0",inplace=True)
female_data_20s = pd.read_csv(f_20s_path)
female_data_20s.drop(columns="Unnamed: 0",inplace=True)
male_data_30s = pd.read_csv(m_30s_path)
male_data_30s.drop(columns="Unnamed: 0",inplace=True)
female_data_30s = pd.read_csv(f_30s_path)
female_data_30s.drop(columns="Unnamed: 0",inplace=True)
```

```python
data_20s = pd.merge(male_data_20s,female_data_20s,how="outer",on="19-29 음식")
data_30s = pd.merge(male_data_30s,female_data_30s,how="outer",on="30-49 음식")
```

```python
from numpy import NaN
inf = float("inf")
# 순위 차이 계산하기 (20대)
rank_dif = []
abs_rank_dif = []
for i in range(len(data_20s)):
  mr = data_20s.iloc[i,0]
  fr = data_20s.iloc[i,5]

  rank_dif.append(mr-fr)
  abs_rank_dif.append(abs(mr-fr))

data_20s["남자-여자 순위 차이"] = rank_dif
data_20s["abs(남자-여자 순위 차이)"] = abs_rank_dif
data_20s
```

```python
data_20s = data_20s.loc[:,["순위_x","순위_y","19-29 음식","19-29 비율(%)_x","19-29 비율(%)_y","남자-여자 순위 차이","abs(남자-여자 순위 차이)"]]
data_20s.rename(columns={"순위_x":"남자 순위","순위_y":"여자 순위","19-29 음식":"음식","19-29 비율(%)_x":"남자 비율","19-29 비율(%)_y":"여자 비율"},inplace=True)
```

```python
# 순위 차이 계산하기 (30대)
rank_dif = []
abs_rank_dif = []
for i in range(len(data_30s)):
  mr = data_30s.iloc[i,0]
  fr = data_30s.iloc[i,5]

  rank_dif.append(mr-fr)
  abs_rank_dif.append(abs(mr-fr))

data_30s["남자-여자 순위 차이"] = rank_dif
data_30s["abs(남자-여자 순위 차이)"] = abs_rank_dif
data_30s
```

```python
# 박스 플롯 그려보기 
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

data = np.array(common_data.iloc[:,-1])

fig,axes = plt.subplots()

box = axes.boxplot(data)
axes.set_ylabel("rank")
plt.title("rank-diff boxplot")
plt.yticks(np.arange(0,60,10))
plt.show()
```

```python
whiskers = [item.get_ydata() for item in box['whiskers']]
medians = [item.get_ydata() for item in box['medians']]
fliers = [item.get_ydata() for item in box['fliers']]
avg = data.mean()
print('whiskers:', whiskers) # 1사분위 수, 3사분위 수
print('medians:', medians) # 중앙값
print('fliers:', fliers) # 이상치 
print("average:",avg) # 평균 
```

```python
# 히스토그램 
Q1 = np.quantile(data, 0.25)
Q2 = np.quantile(data, 0.5)
Q3 = np.quantile(data, 0.75)
avg = np.mean(data)

fig,axes = plt.subplots()
axes.hist(data)
axes.set_xlabel("rank-diff")
axes.set_ylabel("num")
plt.xticks(np.arange(0,60,5))
plt.axvline(Q1, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

```python
print("평균 :",avg)
print("중앙값 :",Q2)
print("1사분위수 :",Q1)
```
